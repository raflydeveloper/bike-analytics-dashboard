## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
- Bagaimana distribusi jumlah penyewaan sepeda pada musim hujan?
```

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
    day_df[column] = pd.to_datetime(day_df[column])

# Filter data hanya untuk musim hujan (season = 4)
rainy_df = day_df[day_df['season'] == 4]

# Resample harian (atau bisa juga 'W' untuk mingguan jika ingin lebih smooth)
daily_rainy_df = rainy_df.resample(rule='D', on='dteday').agg({
    "cnt": "sum"
}).reset_index()

# Rename kolom untuk konsistensi
daily_rainy_df.rename(columns={
    "dteday": "date",
    "cnt": "total_rentals"
}, inplace=True)

# Plot mirip dengan grafik harga saham (time series line plot)
plt.figure(figsize=(12, 5))
plt.plot(daily_rainy_df['date'], daily_rainy_df['total_rentals'], color='red')
plt.xlabel('Tanggal', size=15)
plt.ylabel('Jumlah Penyewaan', size=15)
plt.title('Tren Jumlah Penyewaan Sepeda Selama Musim Hujan', size=16)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
#- Bagaimana pola jumlah penyewaan sepeda di setiap musim, dan musim mana yang memiliki jumlah penyewaan tertinggi?
```

```python
# Mengkonversi kolom tanggal menjadi datetime
datetime_columns = ["dteday"]
for column in datetime_columns:
    hour_df[column] = pd.to_datetime(hour_df[column])

# Mengubah kolom kategorikal menjadi tipe 'category' untuk efisiensi memori
categorical_cols = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
hour_df[categorical_cols] = hour_df[categorical_cols].astype('category')

# Menghitung total penyewaan sepeda berdasarkan musim ('season') dengan menambahkan observed=False
seasonal_rentals_df = hour_df.groupby('season', observed=False).agg({
    'cnt': 'sum'
}).reset_index()

# Menambahkan nama musim untuk interpretasi yang lebih mudah (jika diperlukan)
season_mapping = {1: 'Musim Semi', 2: 'Musim Panas', 3: 'Musim Gugur', 4: 'Musim Hujan'}
seasonal_rentals_df['season_name'] = seasonal_rentals_df['season'].map(season_mapping)

# Mengurutkan berdasarkan jumlah penyewaan tertinggi
seasonal_rentals_df = seasonal_rentals_df.sort_values(by='cnt', ascending=False)

# Plot distribusi jumlah penyewaan sepeda berdasarkan musim
plt.figure(figsize=(10, 5))
plt.bar(seasonal_rentals_df['season_name'], seasonal_rentals_df['cnt'], color="#72BCD4")
plt.title("Jumlah Penyewaan Sepeda Berdasarkan Musim", fontsize=16)
plt.xlabel("Musim")
plt.ylabel("Jumlah Penyewaan")
plt.tight_layout()
plt.show()

# Menampilkan musim dengan jumlah penyewaan tertinggi
highest_rentals_season = seasonal_rentals_df.iloc[0]
print(f"Musim dengan jumlah penyewaan tertinggi adalah {highest_rentals_season['season_name']} dengan total penyewaan {highest_rentals_season['cnt']}.")
```
